```python
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
import pandas as pd

# Data SMILES
data = filtered_ligands[['canonical_smiles', 'pchembl_value']]

df = pd.DataFrame(data)

# Fungsi untuk mendapatkan urutan asam amino dari SMILES
def smiles_to_sequence(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            sequence = rdMolDescriptors.CalcMolFormula(mol)  # Misalnya, dapatkan formula molekul
            return sequence  # Placeholder: gunakan algoritma lain jika spesifik ke peptide
        else:
            return None
    except:
        return None

# Tambahkan kolom urutan asam amino
df['amino_acid_sequence'] = df['canonical_smiles'].apply(smiles_to_sequence)


# Simpan ke CSV
df.to_csv("ligands_with_amino_acid_sequence.csv", index=False)
# Lihat hasil
print(df)
# Hapus duplikasi berdasarkan kolom canonical_smiles
df = df.drop_duplicates(subset='canonical_smiles')
print(f"Jumlah data setelah menghapus duplikasi: {len(df)}")
```

# Filtering Using RNN / Transformers

Install Torch

```python
!pip install transformers torch datasets
```

Prepare Data

```python
import pandas as pd

# Contoh dataset
data = {
    "canonical_smiles": [
        "CC(C[C@H](N)C(=O)O)N=C(N)N",  # Arginine
        "C(C(=O)O)N",  # Glycine
        "C(C(C(=O)O)N)C(=O)O",  # Aspartate
        "C1CC[C@@H](C(=O)N[C@@H](CC(=O)O)C(=O)N[C@@H](CCC(=N)N)C(=O)N1)",  # Cyclic RGD
        "C(CC(=O)O)C(=O)O"  # Molekul lain
    ],
    "label": [1, 0, 0, 1, 0]  # 1 = Mengandung motif RGD/cRGD
}
df = pd.DataFrame(data)
```

Tokenisasi

```python
from transformers import AutoTokenizer

# Muat tokenizer ChemBERTa
tokenizer = AutoTokenizer.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")

# Tokenisasi data SMILES
def tokenize_data(data):
    return tokenizer(data["canonical_smiles"], padding=True, truncation=True, max_length=128)

# Tokenize dataset
tokenized_data = df.apply(tokenize_data, axis=1)
```

Transformer

```python
from transformers import AutoModelForSequenceClassification

# Muat model ChemBERTa untuk klasifikasi
model = AutoModelForSequenceClassification.from_pretrained(
    "seyonec/ChemBERTa-zinc-base-v1",
    num_labels=2  # 2 label: mengandung RGD atau tidak
)
```

Membagi dataset untuk training

```python
from sklearn.model_selection import train_test_split
import torch

# Bagi dataset
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Siapkan input untuk model
train_tokens = tokenizer(list(train_data["canonical_smiles"]), padding=True, truncation=True, return_tensors="pt", max_length=128)
test_tokens = tokenizer(list(test_data["canonical_smiles"]), padding=True, truncation=True, return_tensors="pt", max_length=128)

# Siapkan label
train_labels = torch.tensor(train_data["label"].values)
test_labels = torch.tensor(test_data["label"].values)
```

Fine Tuning

```python
from torch.utils.data import DataLoader, TensorDataset
from torch.optim import AdamW
from transformers import get_scheduler

# Dataset dan DataLoader
train_dataset = TensorDataset(train_tokens["input_ids"], train_tokens["attention_mask"], train_labels)
train_dataloader = DataLoader(train_dataset, batch_size=16)

# Optimizer dan scheduler
optimizer = AdamW(model.parameters(), lr=5e-5)
num_training_steps = len(train_dataloader) * 5  # 5 epochs
lr_scheduler = get_scheduler("linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps)

# Gunakan GPU jika tersedia
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
model.to(device)

# Latih model
from tqdm import tqdm

model.train()
for epoch in range(5):  # 5 epochs
    loop = tqdm(train_dataloader, leave=True)
    for batch in loop:
        input_ids, attention_mask, labels = [x.to(device) for x in batch]

        # Forward pass
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        logits = outputs.logits

        # Backward pass
        loss.backward()
        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()

        # Tampilkan loss
        loop.set_description(f"Epoch {epoch}")
        loop.set_postfix(loss=loss.item())
```

Evaluasi Model

```python
from sklearn.metrics import accuracy_score, classification_report

# Siapkan DataLoader untuk testing
test_dataset = TensorDataset(test_tokens["input_ids"], test_tokens["attention_mask"], test_labels)
test_dataloader = DataLoader(test_dataset, batch_size=16)

# Evaluasi
model.eval()
predictions, true_labels = [], []

with torch.no_grad():
    for batch in test_dataloader:
        input_ids, attention_mask, labels = [x.to(device) for x in batch]

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        predictions.extend(torch.argmax(logits, dim=1).tolist())
        true_labels.extend(labels.tolist())

# Hasil evaluasi
print("Accuracy:", accuracy_score(true_labels, predictions))
print("Classification Report:\n", classification_report(true_labels, predictions))
```

Prediksi SMILES baru

```python
new_smiles = ["C(C(=O)O)N"]  # Contoh SMILES baru
new_tokens = tokenizer(new_smiles, padding=True, truncation=True, return_tensors="pt", max_length=128).to(device)

model.eval()
with torch.no_grad():
    logits = model(**new_tokens).logits
    predictions = torch.argmax(logits, dim=1).tolist()

print("Prediksi:", predictions)  # 1 = Mengandung motif, 0 = Tidak
```

```python
import pandas as pd

# Baca file CSV yang diunggah
file_path = "filtered_ligands.csv"  # Ganti dengan nama file yang diunggah
df = pd.read_csv(file_path)

# Tampilkan beberapa baris data untuk verifikasi
print(df.head())
```

```python
# Periksa apakah kolom `canonical_smiles` ada
if 'canonical_smiles' in df.columns:
    print("Kolom 'canonical_smiles' ditemukan!")
else:
    print("Kolom 'canonical_smiles' tidak ditemukan! Periksa nama kolom di dataset Anda.")
```

Tokenisasi SMILES

```python
from transformers import AutoTokenizer

# Muat tokenizer dari model pra-terlatih (misalnya ChemBERTa)
tokenizer = AutoTokenizer.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")

# Tokenisasi kolom SMILES
tokens = tokenizer(
    list(df["canonical_smiles"]),  # Pastikan kolom ini sesuai dengan dataset Anda
    padding=True,
    truncation=True,
    return_tensors="pt",
    max_length=128
)
```

Prediksi Dataset

```python
from transformers import AutoModelForSequenceClassification
import torch

# Muat model terlatih Anda
model = AutoModelForSequenceClassification.from_pretrained(
    "seyonec/ChemBERTa-zinc-base-v1",
    num_labels=2  # Dua label: 1 = mengandung motif, 0 = tidak
)

# Gunakan GPU jika tersedia
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Prediksi dataset
model.eval()
with torch.no_grad():
    outputs = model(
        input_ids=tokens["input_ids"].to(device),
        attention_mask=tokens["attention_mask"].to(device)
    )
    logits = outputs.logits
    predictions = torch.argmax(logits, dim=1).cpu().tolist()  # Prediksi: 1 atau 0

# Tambahkan kolom prediksi ke dataset
df["prediction"] = predictions
```

Filterinfg berdasarkan Prediksi

```python
# Filter dataset berdasarkan prediksi
filtered_df = df[df["prediction"] == 1]

# Simpan dataset yang sudah difilter ke file baru
output_path = "/content/filtered_rgd_dataset.csv"
filtered_df.to_csv(output_path, index=False)

# Tampilkan hasil
print(f"Jumlah data yang mengandung RGD/cRGD: {len(filtered_df)}")
print(filtered_df.head())
```
